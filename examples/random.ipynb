# Comparative between *iarray* and *numpy* random methods

```python
import numpy as np
import iarray as ia
import matplotlib.pyplot as plt
from scipy import stats
```

```python
cfg = ia.Config()
ctx = ia.Context(cfg)
r_ctx = ia.RandomContext(ctx)
size = 1000000
shape = [size]
```

```python
a = ia.random_rand(ctx, r_ctx, shape)
b = ia.iarray2numpy(ctx, a)
c = np.random.rand(size)
d = np.random.randn(size)

plt.hist(b, bins='auto', density=True, histtype="step")
plt.hist(c, bins='auto', density=True, histtype="step")
plt.show()
plt.close()
```

```python
a = ia.random_randn(ctx, r_ctx, shape)
b = ia.iarray2numpy(ctx, a)
c = np.random.randn(size)

plt.hist(b, bins='auto', density=True, histtype="step")
plt.hist(c, bins='auto', density=True, histtype="step")
plt.show()
plt.close()
```

```python
alpha = 2
beta = 6

a = ia.random_beta(ctx, r_ctx, alpha, beta, shape)
b = ia.iarray2numpy(ctx, a)
c = np.random.beta(alpha, beta, size)

plt.hist(b, bins='auto', density=True, histtype="step")
plt.hist(c, bins='auto', density=True, histtype="step")
plt.show()
plt.close()
```

```python
mu = 1
sigma = 0.5

a = ia.random_lognormal(ctx, r_ctx, mu, sigma, shape)
b = ia.iarray2numpy(ctx, a)
c = np.random.lognormal(mu, sigma, size)

plt.hist(b, bins='auto', density=True, histtype="step")
plt.hist(c, bins='auto', density=True, histtype="step")
plt.show()
plt.close()
```
